# Algorithm Performance Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Search Algorithm Choice


The game presents particular characteristics that influenced the choice of algorithm. The most critical aspect in Pipe Mania is to find a solution quickly. Unlike some optimization problems where the shortest path or minimal cost are the main objectives, these factors are not relevant in this context.

We compared four different search algorithms, analyzing their memory usage and execution time. Both the greedy search and the A* search performed the best, with the greedy search outperforming the A* search.

The Greedy algorithm operates based on the concept of making the locally optimal choice at each step of the search process. At each step, the Greedy algorithm selects the option that appears to be the most advantageous without considering the future impact of that choice. It is characterized as being complete within the context of the Pipe Mania problem because it will never enter cycles, though it is not an optimal algorithm.


```python
# Read the data from the file
data = pd.read_csv('time_comparison.csv')
data
```

```python
# Read the CSV file
df = pd.read_csv('time_comparison.csv', skiprows=[18])

# Extract data for plotting
num_pipes = df['num_pipes']
time_dfs = df['time_dfs']
time_bfs = df['time_bfs']
time_astar = df['time_astar']
time_greedy = df['time_greedy']

# Create a plot
plt.figure(figsize=(10, 6))

# Plotting the data
plt.plot(num_pipes, time_dfs, marker='o', label='DFS')
plt.plot(num_pipes, time_bfs, marker='s', label='BFS')
plt.plot(num_pipes, time_astar, marker='^', label='A* Search')
plt.plot(num_pipes, time_greedy, marker='d', label='Greedy Search')


# Adding titles and labels
plt.title('Algorithm Performance Comparison')
plt.xlabel('Number of Pipes')
plt.ylabel('Execution Time (s)')
plt.legend().set_title('Search Algorithm')


# Adding memory usage information
textstr = '\n'.join((
    f'DFS: {"0.918 MB"}',
    f'BFS: {"1.254 MB"}',
    f'A* Search: {"0.9675 MB"}',
    f'Greedy Search: {"1.028 MB"}'
))

# These are matplotlib.patch.Patch properties
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)

# Place the text box in the plot
plt.text(0.62, 0.95, textstr, transform=plt.gca().transAxes, fontsize=12,
         verticalalignment='top', bbox=props)

# Display the plot
plt.show()
```

Now we will compare the DFS algorithm with the greedy search algorithm. Both of these algorithms are designed to find solutions efficiently, but they operate under different principles.

```python
plt.plot(num_pipes, time_dfs, marker='^', label='DFS')
plt.plot(num_pipes, time_greedy, marker='d', label='Greedy Search')


# Adding titles and labels
plt.title('Algorithm Performance Comparison')
plt.xlabel('Number of Pipes')
plt.ylabel('Execution Time (s)')
plt.legend().set_title('Search Algorithm')


# Adding memory usage information
textstr = '\n'.join((
    f'DFS: {"0.918 MB"}',
    f'Greedy Search: {"1.028 MB"}'
))

# These are matplotlib.patch.Patch properties
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)

# Place the text box in the plot
plt.text(0.4, 0.95, textstr, transform=plt.gca().transAxes, fontsize=12,
         verticalalignment='top', bbox=props)

# Display the plot
plt.show()
```

## Heuristic Choice Comparison

The choice of heuristic is crucial for the performance of the Greedy algorithm. In this context, two distinct heuristics were tested.
It was concluded that the first heuristic, which prioritizes states with fewer pieces left to explore, achieved better results. This is because it encourages the algorithm to search new areas of the solution space. In both the first and second heuristics, we ensured that invalid states always have a higher heuristic value. The following table and graph allows us to conclude the first heuristic is better.

```python
# Read the CSV file
df = pd.read_csv('heuristic_comparison.csv', skiprows=[18])
df
```

```python
# Read the CSV file
df = pd.read_csv('heuristic_comparison.csv', skiprows=[18])

# Extract data for plotting
num_pipes = df['num_pipes']
greedy_h1 = df['greedy_h1']
greedy_h2 = df['greedy_h2']

# Create a plot
plt.figure(figsize=(10, 6))

# Plotting the data
plt.plot(num_pipes, greedy_h1, marker='o', label='Greedy H1')
plt.plot(num_pipes, greedy_h2, marker='s', label='Greedy H2')

# Adding titles and labels
plt.title('Algorithm Performance Comparison')
plt.xlabel('Number of Pipes')
plt.ylabel('Execution Time (s)')
plt.legend().set_title('Heuristic Choice')


# Adding memory usage information
textstr = '\n'.join((
    f'Greedy H1: {"1.028 MB"}',
    f'Greedy H2: {"1.502 MB"}'
))

# These are matplotlib.patch.Patch properties
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)

# Place the text box in the plot
plt.text(0.3, 0.95, textstr, transform=plt.gca().transAxes, fontsize=12,
         verticalalignment='top', bbox=props)

# Display the plot
plt.show()
```

## Board Exploration Method Comparison

Now let's compare board exploring methods when applied to the same search algorithm with the same heuristic. As seen by the graph below, exploring the board diagonally offers better performance than exploring the board from the center or exploring the board in layers (shells).

```python
# Read the CSV file
df = pd.read_csv('board_exploring_method.csv', skiprows=[18])
df
```

```python
# Read the CSV file
df = pd.read_csv('board_exploring_method.csv', skiprows=[18])

# Extract data for plotting
num_pipes = df['num_pipes']
greedy_center = df['greedy_center']
greedy_diag = df['greedy_diag']
greedy_shells = df['greedy_shells']

# Create a plot
plt.figure(figsize=(10, 6))

# Plotting the data
plt.plot(num_pipes, greedy_center, marker='o', label='Greedy Center')
plt.plot(num_pipes, greedy_diag, marker='s', label='Greedy Diagonal')
plt.plot(num_pipes, greedy_shells, marker='^', label='Greedy Shells')

# Adding titles and labels
plt.title('Algorithm Performance Comparison')
plt.xlabel('Number of Pipes')
plt.ylabel('Execution Time (s)')
plt.legend().set_title('Board Exploring Method')


# Adding memory usage information
textstr = '\n'.join((
    f'Greedy Center: {"0.9598 MB"}',
    f'Greedy Diagonal: {"1.028 MB"}',
    f'Greedy Shells: {"1.834 MB"}'
))

# These are matplotlib.patch.Patch properties
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)

# Place the text box in the plot
plt.text(0.3, 0.95, textstr, transform=plt.gca().transAxes, fontsize=12,
         verticalalignment='top', bbox=props)

# Display the plot
plt.show()
```
